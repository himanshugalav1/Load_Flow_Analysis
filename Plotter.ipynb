```python
import matplotlib.pyplot as plt
import numpy as np

# Load your data
I_anti = np.loadtxt("I_par.csv")
I_par = np.loadtxt("I_anti.csv")

x = np.linspace(-0.2, 0.2, 10)
y2 = (10**6) * I_anti
y1 = (10**6) * I_par

# Configure font settings
plt.rcParams['font.weight'] = 'bold'
plt.rcParams.update({'font.size': 30})

# Create a figure with a white background
fig = plt.figure(figsize=(10, 8))
fig.patch.set_facecolor('white')

# Plot the data and set line properties
cp, = plt.plot(x, y1, label='P', color='b', linewidth=3)
cn, = plt.plot(x, y2, label='AP', color='r', linewidth=3, zorder=3)

# Set labels and title
plt.xlabel('Voltage', fontweight='bold', fontsize=36)
plt.ylabel('Charge Current (uA)', fontweight='bold', fontsize=36)

# Set custom axis limits and tick positions
x_min, x_max = -0.2, 0.2
y_min, y_max = -40, 40
x_ticks = [-0.2, -0.1, 0, 0.1, 0.2]
y_ticks = [-40, -20, 0,20, 40]

plt.xticks(x_ticks)
plt.yticks(y_ticks)
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)

# Create a legend without a box outline
lgd = plt.legend([cp, cn], ['P', 'AP'])
lgd.set_frame_on(False)

# Customize tick parameters and spines
plt.gca().tick_params(axis='both', which='major', width=3)
plt.gca().spines['top'].set_linewidth(2)
plt.gca().spines['bottom'].set_linewidth(2)
plt.gca().spines['left'].set_linewidth(2)
plt.gca().spines['right'].set_linewidth(2)

plt.tight_layout()
plt.savefig('IV.png')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
X = X_
Y1 = Y_
Y2 = Y2_  # Assuming Y2_ is your second dataset

# Create a figure with a white background
fig = plt.figure(figsize=(10, 8))
fig.patch.set_facecolor('white')

# Plot the first dataset and set line properties
ax1 = plt.gca()
ax1.plot(X, Y1, label='SC', linewidth=3, color='b')  # First dataset on the left y-axis
ax1.set_xlabel('Voltage (V)', fontsize=36,fontweight='bold')

# Create a second y-axis on the right side
ax2 = ax1.twinx()
ax2.plot(X, Y2, label='UMC', linewidth=3, color='r',linestyle='dashed')  # Second dataset on the right y-axis

# Set line properties for the second dataset
lines = [ax1.get_lines()[0], ax2.get_lines()[0]]
labels = [line.get_label() for line in lines]
legend = ax1.legend(lines, labels, loc='lower center')

# Set the legend frame to False
legend.set_frame_on(False)

# Customize tick parameters and spines
ax1.tick_params(axis='both', which='major', width=3)
ax1.spines['top'].set_linewidth(2)
ax1.spines['bottom'].set_linewidth(2)
ax1.spines['left'].set_linewidth(2)
ax1.spines['right'].set_linewidth(2)

# Set y-axis limits and spacing for the left y-axis
y1_min = 57# Minimum y-axis limit for Y1
y1_max = 77  # Maximum y-axis limit for Y1
y1_step =5  # Spacing between ticks for Y1

# Generate y-axis ticks for Y1 using NumPy
y1_ticks = np.arange(y1_min, y1_max + y1_step, y1_step)

ax1.set_yticks(y1_ticks)
ax1.set_ylim(y1_min, y1_max)
ax1.set_ylabel('TMR (%) -SC', fontsize=36,fontweight='bold')

# Set y-axis limits and spacing for the right y-axis
y2_min = 180  # Minimum y-axis limit for Y2
y2_max = 260  # Maximum y-axis limit for Y2
y2_step = 20  # Spacing between ticks for Y2

# Generate y-axis ticks for Y2 using NumPy
y2_ticks = np.arange(y2_min, y2_max + y2_step, y2_step)

#ax2.set_yticks(y2_ticks)
ax2.set_ylim(y2_min, 260)
ax2.set_ylabel('TMR (%) -UMC', fontsize=36,fontweight='bold',color='r')
plt.tight_layout()
plt.savefig('TMR.png')
plt.show()
```
